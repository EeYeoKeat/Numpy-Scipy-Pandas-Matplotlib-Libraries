In this section, we use statistical functions from scipy

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#In this section, we use statistical functions from scipy
from scipy import stats
```

In statistics, a **univariate distribution** is a probability distribution of only one random variable. This is in contrast to a multivariate distribution, the probability distribution of a random vector

```
x = np.random.normal(size=100)
x
```

```
sns.distplot(x);
```

`distplot` is flexibly plot a univariate distribution of observations.By default, it will draw a histogram and fit a kernel density estimate (KDE). Based on [wiki](https://en.wikipedia.org/wiki/Kernel_density_estimation), KDE is the non-parametric statistic way to estimate the probability density function (PDF) of a random variable. 


In addition, we can use pandas series to label axis.

```
x = pd.Series(x, name="x variable")
graph = sns.distplot(x)
```

```
car_crash = sns.load_dataset("car_crashes")
car_crash
```

```
speed = car_crash['speeding']
x = pd.Series(speed, name="Speeding")
graph = sns.distplot(x)
```

Add a rug plot, which draws a small vertical tick at each observation

```
sns.distplot(x, kde=False, rug=True);
```
