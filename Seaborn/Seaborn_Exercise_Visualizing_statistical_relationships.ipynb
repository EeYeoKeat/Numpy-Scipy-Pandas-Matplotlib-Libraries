```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

import data set from the seaborn [built-in data set](https://github.com/mwaskom/seaborn-data)

```
flights = sns.load_dataset("flights")
flights
```

By default, `relplot()` is the scatter plot

```
sns.relplot(x="passengers", y="year", data=flights);
```

Unless specifically use `kind="line"`

```
sns.relplot(x="passengers", y="year", kind="line", data=flights);
```

```
sns.relplot(x="year", y="month", hue="passengers", data=flights);
```
