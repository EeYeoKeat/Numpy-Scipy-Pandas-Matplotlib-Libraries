# Categorical scatterplots

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
iris = sns.load_dataset("iris")
iris
```

```
sns.catplot(x="species", y="sepal_length", data=iris);
```

As we can observe the data point position is slightly “jitter” for every species axis. To make it align vertically, `jitter` parameter can be adjusted.

```
sns.catplot(x="species", y="sepal_length", jitter=False, data=iris);
```

The data points are overlapping to align in one straight line. We can filter the data as well.

```
sns.catplot(x="species", y="sepal_length", kind="swarm", data=iris.query("sepal_length > 5"));
```

**Boxplot**

```
sns.catplot(x="species", y="sepal_length", kind="box", data=iris);
```

One of the data point that fall outside the range are displayed independently.

```
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y="species", data=iris, color="c");
```

```
sns.catplot(data=iris, orient="h", kind="box");
```

```
sns.violinplot(x=iris.species, y=iris.sepal_length);
```

```
sns.violinplot(x=iris.species, y=iris.sepal_width);
```

```
# use another data set
tips = sns.load_dataset("tips")
tips
```

```
g = sns.catplot(x="total_bill", y="time", row="sex",
                kind="box", orient="h", height=2, aspect=5,
                data=tips.query("tip > 1"))
g.set(xscale="log");
```
