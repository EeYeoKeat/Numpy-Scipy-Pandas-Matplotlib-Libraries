To visualize a **bivariate distribution** of two variables, `jointplot()` can be used to shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes.

````
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
````

````
iris = sns.load_dataset("iris")
iris
````

````
sepal_length = iris['sepal_length']
sepal_width = iris['sepal_width']
combined_df = pd.concat([sepal_length, sepal_width], axis=1)
df = pd.DataFrame(combined_df, columns=["sepal_length", "sepal_width"])
df
````

````
sns.jointplot(x="sepal_length", y="sepal_width", data=df);
````

````
with sns.axes_style("white"):
    sns.jointplot(x=sepal_length, y=sepal_width, kind="hex", color="k");
````

Tips to change colour. Commands which take color arguments can use several formats to specify the colors. For the basic built-in colors, you can use a single letter.

```
b: blue
g: green
r: red
c: cyan
m: magenta
y: yellow
k: black
w: white
```
For a greater range of colors, We can specify the color using an html hex string.


````
with sns.axes_style("white"):
    sns.jointplot(x=sepal_length, y=sepal_width, kind="hex", color="");
````

````
sns.jointplot(x="sepal_length", y="sepal_width", data=df, kind="kde");
````

We also can draw a two-dimensional kernel density plot with the `kdeplot()` function.

````
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(df.sepal_length, df.sepal_width, ax=ax)
sns.rugplot(df.sepal_length, color="g", ax=ax)
sns.rugplot(df.sepal_width, vertical=True, ax=ax);
````

````
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df.sepal_length, df.sepal_width, cmap=cmap, n_levels=20, shade=True);
````

To plot multiple pairwise bivariate distributions in a dataset, we can use the `pairplot()` function. This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame. By default, it also draws the univariate distribution of each variable on the diagonal Axes

````
sns.pairplot(iris);
````

````
sns.pairplot(iris, hue="species");
````
