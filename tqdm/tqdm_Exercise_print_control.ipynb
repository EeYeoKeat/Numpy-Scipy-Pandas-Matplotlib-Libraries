```
epoch = 1000
batch_size = 256
x = 500

for e in range(epoch):
  pbar = tqdm(
      range(0, x - 1, batch_size), desc = 'train minibatch loop'
    )
  for i in pbar:
    index = min(i + batch_size, x - 1)
    time.sleep(0.01)
  if e % 50 == 0:
    print('\nepoch: %d, i: %f'%(e, i))
```

As a result, the epoch number only printed out when the epoch number multiple of 50. While the `tqdm` is keep running and display the progress bar. In order to avoid printing too much lines in browser, we actually can no need to use `tqdm`.
