# Plot Graph

```
import numpy as np
import matplotlib.pyplot as plt
```

```
# linspace is generate data with three arguments (start point, end point, number amount of point)
x = np.linspace(0,10,10)
```

```
y = np.sin(x)
```

```
#plot sin graph
plt.plot(x,y)
```

```
plt.show()
```

Now we can add on x and y-axis title and also graph title

```
plt.xlabel("Time")
plt.ylabel("Function of time")
plt.title("Sine Function versus Time")
plt.plot(x,y)
```

Since the graph only has 10 points, the line is not smooth. Therefore, we can add on more point to the graph.

```
x = np.linspace(0,10,100)
y = np.sin(x)
plt.xlabel("Time")
plt.ylabel("Function of time")
plt.title("Sine Function versus Time")
plt.plot(x,y)
```
