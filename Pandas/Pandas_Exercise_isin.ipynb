The `isin()` function is used to check each element in the DataFrame is contained in values or not. It helps in selecting rows with having a particular(or Multiple) value in a particular column.

```
import numpy as np
import pandas as pd
import seaborn as sn
```

```
# to make data load easier, we use dataset from seaborn
df = sn.load_dataset('tips')
df
```

```
data = df["sex"].isin(["Male"]) 
  
# displaying data with gender = Male only 
df[data] 
```

# **Multiple parameter Filtering**

```
# filter 1 is to select non-smoker
filter1 = df["smoker"].isin(["No"])

# filer 2 is to select day when it is Friday or Saturday or Sunday
filter2 = df["day"].isin(["Fri", "Sat", "Sun" ]) 
  
# displaying data with both filter applied and mandatory  
df[filter1 & filter2] 
```
